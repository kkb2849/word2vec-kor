```python
import gensim
```

```python
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
def iter_file(file_name):
    with open(file_name, 'r') as f:
        while True:
            line = f.readline()
            if not line:
                break;
            li = line.replace('\n','').split()
            arr = []
            for word in li:
                sp = word.split('_')
                if len(sp) != 1:
                    if sp[1]=='Noun' or sp[1]=='Verb' or sp[1] =='Adjective':
                        arr.append(sp[0])
                else:
                    arr.append(word)
            if len(arr) != 0:
                yield arr
```

```python
filename = '/mnt/data/test_data/namuwiki_stemmed.txt'

size = 300
window = 5
min_count = 20
```

```python
it = iter_file(filename)
for i,line in enumerate(it):
    if i > 5:
        break;
    print(line)
    
```

```python
model = gensim.models.Word2Vec(size=size, window=window, min_count=min_count, workers=16, sg=1, iter=10)
# model = gensim.models.Word2Vec.load('/home/tf_user/test_data/word2vec/trained_namu_sg_' + str(size) + '_' + str(window) + '_' + str(min_count))
it = iter_file(filename)
model.build_vocab(it)
```

```python
it = iter_file(filename)
# model.train(it, total_examples=model.corpus_count, epochs=100, start_alpha=0.025, end_alpha=1e-3)
model.train(it, total_examples=1922928.5, epochs=10)
model.save('/home/tf_user/test_data/word2vec/170906_for20_trained_namu_sg_' + str(size) + '_' + str(window) + '_' + str(min_count))
```

```python
print(model.wv.similarity('이명박', '박근혜'))
print(model.wv.most_similar(positive=['이명박', '여자'], negative=['남자']))
print(model.wv.most_similar(positive=['중국', '서울'], negative=['북경']))
```
