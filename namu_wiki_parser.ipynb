```python
import ijson

input_file_name = '/mnt/data/test_data/namuwiki_161031.json'
output_file_name = '/mnt/data/test_data/parsed_namuwiki.txt'
```

```python
input_file = open(input_file_name, "r")
output_file = open(output_file_name, "w")
    
parser = ijson.parse(input_file)

index = 0
for prefix, event, value in parser:
    if (prefix, event) == ('item.text', 'string'):
        try:
            output_file.write(value)
        except Exception as e:
            print(e)
            
    if index % 5000 == 0:
        print('index: ', index)
    index+=1

input_file.close()
output_file.close()
```
